Now create a vector to test and create the opcodes function

```r
test_1 = c(1,9,10,3,2,3,11,0,99,30,40,50)

opcodes <- function(codes){
    local_codes <- codes
    for (i in seq(1,length(local_codes),4)){
        subset <- local_codes[i:(i+3)] #r index starts at 1 not 0
        pos_1 <- subset[1] #get positions
        pos_2 <- subset[2]
        pos_3 <- subset[3]
        pos_4 <- subset[4]
        if (pos_1==99){ #if it's 99 we break
            break
        } else if (pos_1==1){ #add the positions together if chunk starts with 1
            local_codes[pos_4+1] <- local_codes[pos_2+1]+local_codes[pos_3+1]
        } else if (pos_1==2){ #multiply the positions if the chunk starts with 2        
            local_codes[pos_4+1] <- local_codes[pos_2+1]*local_codes[pos_3+1]
       }
    }
    return(local_codes)
}
```

```r
opcodes(test_1) #testing
```

```r
opcodes(c(1,0,0,0,99)) == c(2,0,0,0,99)#testing
```
